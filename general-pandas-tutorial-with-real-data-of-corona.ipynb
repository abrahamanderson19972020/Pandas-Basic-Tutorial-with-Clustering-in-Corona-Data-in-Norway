# 5. Cleaning Data Before the Use of Algorithm and Implemeting Algorithm: 

```python
#Here I create a new dataframe that I use of for the use of algorithm 
#because I have to drop some columns and need to compare results with the original data
latest_df=df1.drop(["date","date_time","kommune_name","fylke_name"],axis=1)
#I drop ["date","date_time","kommune_name","fylke_name"] columns and assign the rest of the data into new dataframe
```

```python
latest_df.drop("Percentage",axis=1,inplace=True)#I drop also "Percentage column" because it has some missing values
```

```python
latest_df #here is the data ready for the algorithm
```

```python
from sklearn.cluster import KMeans#here I call K Means Clustering Algorithm
#In this data set, I will not make prediction, but I will cluster municipalities in order to explore the performance of K Means
```

```python
model=KMeans(n_clusters=3) 
#Here I creat an instance of the algorithm and select 3 clusters by assigning n_clusters parameter as 3
```

```python
model.fit(latest_df.values)# Here I make the algorithm to fit the data
```

```python
model.cluster_centers_ #These are the cluster centers of the 3 spearate clusters
```

```python
model.labels_# These are the list of the cluster in the entire dataset
```

```python
print(model.labels_.shape) # as we can see the shape of labels match with original data
print(df1.shape)
```

```python
labels=pd.DataFrame(model.labels_,columns=["Labels"]) 
# here I creat a dataframe of labels because I want to add it into the original data in order to compare and see the performance
labels
                
```

```python
df2=pd.concat([df1,labels],axis=1) 
#here I concatenate the original dataframe df1 with the labels dataframe that the algorithm has clustered and create df2 dataframe
```

```python
df2 #as we can see below in the table, we have a Lables column that is added into our original datafram
```

```python
df2.groupby("Labels").mean()["cases"]
# Here we can see that the algorithm assign label 0 if the number of corona cases is around  21
# Here we can see that the algorithm assign label 1 if the number of corona cases is around  5622
# Here we can see that the algorithm assign label 0 if the number of corona cases is around  517
```

```python

df2[df2["Labels"]==1].head(50)
#Here we see that Oslo region has been clustered with 1 label
```

```python
df2[df2["Labels"]==0].head(50) 
#The algorithm has clustered municipalities as cluster 0 if the number of cases are fewer than 50 cases
```

```python
df2[df2["Labels"]==2].head(50) 
#The algorithm has clustered municipalities as cluster 0 if the number of cases are more than 50 cases and fewer than 700
```
